```python
import numpy.random as rn
import matplotlib.pyplot as plt
import math
import numpy as np
import pandas as pd
```

```python
def random_number():
    return rn.rand(2,1)
```

```python
solution_array = np.zeros((4))
plt.figure(figsize=(10, 10))
for i in range(1000):
    point = random_number()[0], random_number()[1]
    distance = math.sqrt(point[0] ** 2 + point[1] ** 2)
    if distance > 1:
        solution_array[1] += 1 # Outside circle
        plt.plot(point[0],point[1], '.b')
    else:
        solution_array[0] += 1 # Inside circle
        plt.plot(point[0],point[1], '.g')
plot_circle = plt.Circle((0,0), 1)
for a in range(100):
    plt.plot(math.cos(math.pi *a/200), math.sin(math.pi * a/200), 'or')
plt.xlim(0,1)
plt.ylim(0,1)
solution_array[2] = solution_array[0] + solution_array[1]
if solution_array[2] != 0:
    solution_array[3] = solution_array[0] / solution_array[2] * 4
solution = 'Inside Outside Total pi-estimation'.split()
info = pd.DataFrame(solution_array, index=solution, columns=['Values'])
info['Values'] = info['Values'].apply(str)
plt.legend(info.reset_index()['index'] + ': ' + info.reset_index()['Values'], bbox_to_anchor=(1, 1),handlelength=0, markerscale=0)
```
